```python
import os
import torch
from PIL import Image
from realesrgan import RealESRGAN

def upscale_for_printing(input_path, output_path, scale=4):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    
    # Load the specialized model for 'Digital Art' (best for cards)
    # Scale=4 turns 1024px into 4096px
    model = RealESRGAN(device, scale=scale)
    model.load_weights("RealESRGAN_x4.pth")

    # Load and process
    img = Image.open(input_path).convert('RGB')
    print(f"Upscaling {input_path}...")
    
    sr_image = model.predict(img)
    
    # Set Metadata to 300 DPI so the printer recognizes the quality
    sr_image.save(output_path, dpi=(300, 300))
    print(f"Print-ready file saved: {output_path}")

if __name__ == "__main__":
    upscale_for_printing("1.png", "FINAL_WEDDING_CARD_300DPI.png")
```
